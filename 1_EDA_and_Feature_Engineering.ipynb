```python
df_2 = pd.read_excel("drive/My Drive/Dataset/Country-Code.xlsx")
df_2.head(2)
```

```python
final_df = pd.merge(df, df_2, on = "Country Code",how= "left" )
final_df.sample(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index 
country_names 
```

```python
country_val = final_df.Country.value_counts().values
country_val 
```

```python
## pie chart--> Top 3 countries
plt.pie(country_val[:3], labels = country_names[:3], autopct = "%1.2f%%")
```

Observation: Zomato maximum records or transactions are from India after that USA then United kingdoms


```python
final_df.columns
```

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index()
ratings
```

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns = {0: "Rating Count"})
ratings
```

**Observation**


1.   When Rating is between 4.5 to 4.9----> Excellent
2.   When Rating is 4.0 to 3.4----> Very good
3.   When Rating is between 3.5 to 3.9----> Good
4.   When Rating is between 3.0 to 3.4----> Average
5.   When Rating is between 2.5 to 2.9---->Poor 




```python
ratings.head(3)
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"] = (12,6)
sns.barplot(x = "Aggregate rating", y = "Rating Count", data = ratings)
```

```python
sns.barplot(x = "Aggregate rating", y = "Rating Count", hue = "Rating color", data = ratings, palette = ["white", "red", "orange", "yellow", "green", "green"]), 
```

**Observation**


1.   Not Rated count is very high
2.   Maximum number of Rating are between 2.5 to 3.4



```python
## count plot
sns.countplot(x = "Rating color",data = ratings, palette = ["white", "red", "orange", "yellow", "green", "green"])
```
